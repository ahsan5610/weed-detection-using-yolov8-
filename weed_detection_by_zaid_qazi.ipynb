```python
# train_yolov8.py  (or run in a notebook cell)
from ultralytics import YOLO
import os

# ---------- CONFIG ----------
DATA_YAML = "/content/CottonWeedYolo-260/data.yaml"   # YOUR path
MODEL_WEIGHTS = "yolov8s.pt"   # start with yolov8s; change to yolov8n for tiny GPU
EPOCHS = 50
IMG_SIZE = 640
BATCH = 16            # reduce if OOM
DEVICE = 0            # 0 for first GPU, or "cpu"
PROJECT = "runs/train"
RUN_NAME = "cotton_weeds_custom"
WORKERS = 4
# ----------------------------

# Create model from pretrained weights
model = YOLO(MODEL_WEIGHTS)

# Train
model.train(
    data=DATA_YAML,
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH,
    device=DEVICE,
    project=PROJECT,
    name=RUN_NAME,
    workers=WORKERS,
    exist_ok=False    # set True to overwrite previous run with same name
)
```

```python
from ultralytics import YOLO

model = YOLO("/content/runs/train/cotton_weeds_custom/weights/best.pt")

results = model.val()

# Per-class metrics
precision = results.box.p
recall = results.box.r
map50 = results.box.map50  # treat as accuracy-like
```

```python
class_names = model.names
labels = [class_names[i] for i in range(len(precision))]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,5))
plt.bar(labels, precision)
plt.title("Precision per Class")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Precision")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.bar(labels, recall)
plt.title("Recall per Class")
plt.xticks(rotation=45, ha='right')
plt.ylabel("Recall")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.bar(labels, map50)
plt.title("mAP50 per Class")
plt.xticks(rotation=45, ha='right')
plt.ylabel("mAP50")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

results_path = "/content/runs/train/cotton_weeds_custom/results.csv"
df = pd.read_csv(results_path)

df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(df["epoch"], df["train/box_loss"], label="Train Box Loss")
plt.plot(df["epoch"], df["train/cls_loss"], label="Train Class Loss")
plt.plot(df["epoch"], df["val/box_loss"], label="Val Box Loss")
plt.plot(df["epoch"], df["val/cls_loss"], label="Val Class Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["epoch"], df["metrics/precision(B)"], label="Precision")
plt.plot(df["epoch"], df["metrics/recall(B)"], label="Recall")
plt.plot(df["epoch"], df["metrics/mAP50(B)"], label="mAP50")
plt.plot(df["epoch"], df["metrics/mAP50-95(B)"], label="mAP50-95")
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Precision, Recall, mAP Over Epochs")
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
import streamlit as st
from ultralytics import YOLO
from PIL import Image
import numpy as np

# Load model
model = YOLO("/content/runs/train/cotton_weeds_custom/weights/best.pt")

st.title("Cotton Weed Detection")

uploaded_file = st.file_uploader("Upload an image", type=["jpg", "jpeg", "png"])

if uploaded_file:
    img = Image.open(uploaded_file)
    st.image(img, caption="Uploaded Image", use_column_width=True)

    # Run detection
    results = model.predict(np.array(img))

    # Show result image
    result_img = results[0].plot()
    st.image(result_img, caption="Detected Weeds", use_column_width=True)

    # Show detected classes
    st.subheader("Detected Classes:")
    detected_classes = results[0].boxes.cls.cpu().numpy()
    names = model.names

    if len(detected_classes) == 0:
        st.write("No weeds detected âœ…")
    else:
        for c in detected_classes:
            st.write(f"- {names[int(c)]}")
```
